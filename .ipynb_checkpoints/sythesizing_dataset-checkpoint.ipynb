```python
all_tags = [
    'beach', 'hiking', 'cultural', 'historical', 'culinary', 'family',
    'adventure', 'photography', 'shopping', 'nature', 'free', 'solo',
    'religious', 'romantic', 'wildlife', 'resort', 'camping', 'art',
    'traditional', 'modern', 'museum', 'island', 'temple', 'mountain',
    'city', 'village', 'festival', 'local', 'luxury', 'budget',
    'spiritual', 'heritage', 'garden', 'forest', 'coastal', 'sunset',
    'sunrise', 'extreme', 'underrated', 'popular', 'eco-friendly'
]

# Function to randomly assign 5 unique tags
def assign_random_tags():
    return ', '.join(random.sample(all_tags, 5))

# Apply to DataFrame
place_df['Tags'] = place_df['Description'].apply(lambda x: assign_random_tags())
```

## Synthesize Click Log

```python
def generate_click_history(users, tourism_df, max_clicks_per_user=20):
    click_history = []
    sources = ['homepage', 'recommendations', 'external_ad', 'search']
    
    for _, user in users.iterrows():
        user_id = user['User_Id']
        city = user['Location']
        user_city_data = tourism_df[tourism_df['City'] == city]
        
        # Allow some clicks outside user's city (20% probability)
        if np.random.rand() < 0.8 and not user_city_data.empty:
            city_places = user_city_data
        else:
            city_places = tourism_df
        
        # Determine number of clicks based on user age (younger users click more)
        base_clicks = min(max_clicks_per_user, max(5, int(user['Age']/5)))
        num_clicks = np.random.randint(base_clicks-3, base_clicks+4)
        
        for _ in range(num_clicks):
            # Select place with probability weighted by rating and popularity
            weights = (city_places['Rating']/5) * np.log1p(city_places['Rating_Count'])
            weights = weights.fillna(0.1).values + 0.1  # Add small base probability
            weights /= weights.sum()
            
            place = city_places.sample(n=1, weights=weights).iloc[0]
            
            # Generate timestamp within last 6 months
            click_time = datetime.now() - timedelta(
                days=np.random.randint(1, 180),
                hours=np.random.randint(0, 24),
                minutes=np.random.randint(0, 60)
            )
            
            click_history.append({
                'User_Id': user_id,
                'Place_Id': place['Place_Id'],
                'Click_Timestamp': click_time.strftime("%Y-%m-%d %H:%M:%S"),
                'Source': np.random.choice(sources, p=[0.3, 0.4, 0.2, 0.1])
            })
    
    return pd.DataFrame(click_history)
```

```python
click_history_df = generate_click_history(user_df, place_df)
```

## Synthesize Search History

```python
def generate_search_history(users, tourism_df, max_searches_per_user=15):
    search_history = []
    all_categories = tourism_df['Category'].unique()
    all_cities = tourism_df['City'].unique()
    place_names = tourism_df['Place_Name'].tolist()
    
    search_templates = [
        "Best {category}",
        "{place_name} reviews",
        "Things to do in {city}",
        "How to get to {place_name}",
        "{category} near me",
        "{place_name} entrance fee",
        "Top rated {category}",
        "Is {place_name} kid-friendly?",
        "Photos of {place_name}",
        "{city} tourist attractions"
    ]
    
    for _, user in users.iterrows():
        user_id = user['User_Id']
        num_searches = np.random.randint(5, max_searches_per_user)
        
        for _ in range(num_searches):
            # Randomly select elements from tourism data
            random_place = tourism_df.sample(1).iloc[0]
            random_category = np.random.choice(all_categories)
            random_city = np.random.choice(all_cities)
            
            # Generate search timestamp within last year
            search_time = datetime.now() - timedelta(
                days=np.random.randint(1, 365),
                hours=np.random.randint(0, 24),
                minutes=np.random.randint(0, 60)
            )
            
            # Create search query using template
            template = np.random.choice(search_templates)
            try:
                query = template.format(
                    category=random_category.lower(),
                    place_name=random_place['Place_Name'],
                    city=random_city
                )
            except KeyError:
                query = template  # Fallback for templates without placeholders
                
            search_history.append({
                'User_Id': user_id,
                'Search_Query': query,
                'Search_Timestamp': search_time.strftime("%Y-%m-%d %H:%M:%S"),
                'Implied_Place_Id': random_place['Place_Id'] if np.random.rand() < 0.6 else None
            })
    
    return pd.DataFrame(search_history)
```

```python
search_history_df = generate_search_history(user_df, place_df)
```

## Synthesize User Features

```python
def enhance_user_features(users_df, tourism_df):
    fake = Faker('id_ID')
    categories = tourism_df['Category'].unique()
    all_tags = ['beach', 'hiking', 'cultural', 'historical', 'culinary', 
               'family', 'adventure', 'photography', 'shopping', 'nature', 'free', 'solo'
               'religious', 'nature']
    
    # Create city coordinates dictionary with proper parsing
    city_coords = {}
    for city, coord_str in tourism_df.groupby('City')['Coordinate'].first().items():
        try:
            coord_dict = ast.literal_eval(coord_str)
            city_coords[city] = (float(coord_dict['lat']), float(coord_dict['lng']))
        except:
            city_coords[city] = (-6.2088, 106.8456)  # Jakarta default
    
    enhanced_users = []
    
    for _, user in users_df.iterrows():
        # Generate coordinates based on location
        base_lat, base_lon = city_coords.get(user['Location'], (-6.2088, 106.8456))
        user_lat = base_lat + np.random.uniform(-0.1, 0.1)
        user_lon = base_lon + np.random.uniform(-0.1, 0.1)
        user_coord = f"{user_lat:.6f},{user_lon:.6f}"
        
        # Generate visited places (3-15% of total places)
        num_visited = np.random.randint(3, len(tourism_df)//7)
        visited = tourism_df.sample(num_visited)['Place_Id'].tolist()
        
        # Determine price preference based on age and visited places
        avg_spent = tourism_df[tourism_df['Place_Id'].isin(visited)]['Price'].mean() if visited else 0
        price_pref = np.clip(int(avg_spent + np.random.normal(0, 50000)), 50000, 500000)
        
        # Generate interest tags (3-5 tags)
        tags = np.random.choice(
            all_tags, 
            size=np.random.randint(3,6), 
            p=[0.15,0.1,0.2,0.1,0.15,0.1,0.05,0.05,0.05,0.05]
        ).tolist()
        
        # Determine preferred categories from visited places
        if visited:
            visited_cats = tourism_df[tourism_df['Place_Id'].isin(visited)]['Category']
            pref_cats = visited_cats.value_counts().head(2).index.tolist()
        else:
            pref_cats = np.random.choice(categories, 2, replace=False).tolist()
        
        # Generate minimum rating preference
        min_rating = max(3.0, np.random.normal(4.0, 0.5))
        
        enhanced_users.append({
            'User_Id': user['User_Id'],
            'Location': user['Location'],
            'Age': user['Age'],
            'Coordinates': user_coord,
            'Visited_Places': visited,
            'Price_Range': f"{price_pref-50000}-{price_pref+50000}",
            'Interest_Tags': tags,
            'Preferred_Categories': pref_cats,
            'Min_Rating': round(min_rating, 1)
        })
    
    return pd.DataFrame(enhanced_users)
```

```python
enhanced_user_df = enhance_user_features(user_df, place_df)
```
